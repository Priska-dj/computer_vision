```python
%matplotlib inline
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.transform import integral_image
from skimage.feature import haar_like_feature
```

```python
# Convert BGR to RGB
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#load test image
img = cv2.imread('./img/3.jpg')

#convert the test image to gray image as opencv face detector expects gray images
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the image
plt.imshow(gray_img, cmap='gray')
```

```python
img1 = cv2.resize(gray_img,(20,20)) 
img1 #ditampilkan dalam bentuk matrix
```

```python
def integral_image_manual(image):
    hasil = np.array([[0 for i in range(20)] for j in range(20)])
    for i in range(20):
        for j in range(20):
            #print(i,j)
            for x in range(i+1):
                for y in range(j+1):
                    #print(x,y)
                    hasil[i][j] = hasil[i][j] + image[x][y]
                    
    return hasil
```

```python
integral_image_manual(img1)
```

```python
#load cascade classifier training file for haarcascade
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')


#let's detect multiscale (some images may be closer to camera than others) images
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.2, minNeighbors=5);

#print the number of faces found
print('Faces found: ', len(faces))
```

```python
#go over list of faces and draw them as rectangles on original colored img
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 4)

#conver image to RGB and show image
plt.imshow(convertToRGB(img))
plt.show()
```
